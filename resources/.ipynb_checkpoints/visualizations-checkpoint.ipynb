```python
from branca.element import MacroElement
from jinja2 import Template

class BindColormap(MacroElement):
    """Binds a colormap to a given layer.

    Parameters
    ----------
    colormap : branca.colormap.ColorMap
        The colormap to bind.
    """
    def __init__(self, layer, colormap):
        super(BindColormap, self).__init__()
        self.layer = layer
        self.colormap = colormap
        self._template = Template(u"""
        {% macro script(this, kwargs) %}
            {{this.colormap.get_name()}}.svg[0][0].style.display = 'block';
            {{this._parent.get_name()}}.on('overlayadd', function (eventLayer) {
                if (eventLayer.layer == {{this.layer.get_name()}}) {
                    {{this.colormap.get_name()}}.svg[0][0].style.display = 'block';
                }});
            {{this._parent.get_name()}}.on('overlayremove', function (eventLayer) {
                if (eventLayer.layer == {{this.layer.get_name()}}) {
                    {{this.colormap.get_name()}}.svg[0][0].style.display = 'none';
                }});
        {% endmacro %}
        """)  # noqa

# This will remove the existing legend from a choropleth map
def RemoveLegend(c):
    for key in c._children:
        if key.startswith('color_map'):
            del(c._children[key])
```

```python
def GetColors(N, map_name='rainbow'):
    '''Function returns a list of N colors from a matplotlib colormap
            Input: N = number of colors, and map_name = name of a matplotlib colormap
    
            For a list of available colormaps: 
                https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html
    '''
    import matplotlib
    cmap = matplotlib.cm.get_cmap(name=map_name)
    n = np.linspace(0, 1, N)
    colors = cmap(n)
    
    # Return as a tuple of tuples
    colors = tuple(map(tuple, colors))
    return colors
```

```python
states = 'gz_2010_us_040_00_500k.json'

def MakeChoropleth(data, column, legend_name, n_bins, colormap='YlGn', column_longname=None):
    
    min_metric = np.floor(min(data[column]))
    max_metric = np.ceil(max(data[column]))
    
    bins = np.linspace(min_metric,max_metric,n_bins+1)
    
    if column_longname is None:
        column_longname=column
    
    c = folium.Choropleth(
                geo_data=states,
                data = data,
                columns=['state',column],
                key_on="feature.properties.NAME",
                fill_color=colormap,
                fill_opacity=1,
                legend_name = legend_name,
                bins = bins,
                line_color='grey',
                line_opacity=0.8,
                line_weight=0.5,
                name = column_longname,
                overlay=True
            )
    # We'll talk about why I'm returning the last two things in a minute...
    return c,min_metric,max_metric
```

```python
# Still working on the Choropleth

salary_tuition['early_career_return_on_total_cost_out_of_state'] = salary_tuition['early_career_pay']/salary_tuition['out_of_state_total']
salary_tuition['mid_career_return_on_total_cost_out_of_state'] = salary_tuition['mid_career_pay']/salary_tuition['out_of_state_total']
salary_tuition['early_career_return_on_total_cost_in_state'] = salary_tuition['early_career_pay']/salary_tuition['in_state_total']
salary_tuition['mid_career_return_on_total_cost_in_state'] = salary_tuition['mid_career_pay']/salary_tuition['in_state_total']

```

```python
m = Map(location = [39,-96],zoom_start=4.2,tiles=None)

n_bins = 7
cmap = 'YlGn'


c1, min_c1, max_c1 = MakeChoropleth(salary_tuition,'early_career_return_on_total_cost_out_of_state', 'early career return',n_bins, column_longname='early career return (out of state)',colormap=cmap)
c2, min_c2, max_c2 = MakeChoropleth(salary_tuition,'mid_career_return_on_total_cost_out_of_state','mid career return', n_bins, column_longname='mid career return (out of state)', colormap=cmap)

c3, min_c3, max_c3 = MakeChoropleth(salary_tuition,'early_career_return_on_total_cost_in_state', 'early career return',n_bins, column_longname='early career return (in state)',colormap=cmap)
c4, min_c4, max_c4 = MakeChoropleth(salary_tuition,'mid_career_return_on_total_cost_in_state','mid career return', n_bins, column_longname='mid career return (in state)', colormap=cmap)



# Remove the existing legends before adding the choropleth layers to the map
RemoveLegend(c1)
RemoveLegend(c2)
RemoveLegend(c3)
RemoveLegend(c4)

# Build some new colorbars
colors1 =  GetColors(7, cmap)


cm1 = branca.colormap.StepColormap(colors, 
                                       vmin=min_c1, vmax=max_c1, 
                                       caption='early career return (out of state)')
cm2 = branca.colormap.StepColormap(colors, 
                                       vmin=min_c2, vmax=max_c2, 
                                       caption='mid career return (out of state)')
cm3 = branca.colormap.StepColormap(colors, 
                                       vmin=min_c3, vmax=max_c3, 
                                       caption='early career return (in state)')
cm4 = branca.colormap.StepColormap(colors, 
                                       vmin=min_c4, vmax=max_c4, 
                                       caption='mid career return (in state)')


m.add_child(c1).add_child(c2).add_child(c3).add_child(c4).add_child(folium.LayerControl())
m.add_child(cm1).add_child(cm2).add_child(cm3).add_child(cm4)
m.add_child(BindColormap(c1, cm1)).add_child(BindColormap(c2, cm2)).add_child(BindColormap(c3, cm3)).add_child(BindColormap(c4, cm4))
```

```python
def AVLnodes(height):
      
    # Base Conditions
    if (height == 1):
        return 1
    elif (height == 2):
        return 2
  
    # Recursive function call
    # for the recurrence relation
    return (1 + AVLnodes(height - 1) + 
                AVLnodes(height - 2))
  
```

```python
def floor_log2(x):
    if x == 1 :
        return 0
    else:
        return 1+floor_log2(x//2)
    
```

```python
def odd_zeroes_1(arr):
    for i in range(1, len(arr), 2):
        arr[i] = 0
```

```python
def odd_zeroes(arr):
    for i in range(len(arr)):
        if arr[i] % 2 == 1:
            arr[i] = 0
```
